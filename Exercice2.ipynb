```python
"""
Exercice 2: Structures de contrôle
Analyse des populations des districts avec if/elif/else et boucles
"""

import pandas as pd
from google.colab import drive

def charger_donnees_districts():
    """Charge les données des districts depuis Google Drive"""
    try:
        drive.mount('/content/drive', force_remount=False)

        base_path = '/content/drive/My Drive/Exercice/DATA/'

        # Charger les données de population
        df = pd.read_excel(base_path + 'Populations_CI_2019.xlsx', sheet_name='CIV_ADM1_POP')

        # Nettoyer les données
        df = df.dropna(subset=['ADM1_FR'])
        df = df[df['ADM1_FR'] != '#adm1+name']
        df['T_TL'] = pd.to_numeric(df['T_TL'], errors='coerce')

        # Prendre les 5 premiers districts
        df_districts = df.head(5)[['ADM1_FR', 'T_TL']].copy()
        df_districts.columns = ['district', 'population']

        return df_districts

    except:
        # Si erreur avec les données réelles, utiliser des exemples
        return pd.DataFrame({
            'district': ['Abidjan', 'Bouaké', 'Daloa', 'Korhogo', 'San Pedro'],
            'population': [5372892, 680529, 478865, 536069, 539635]
        })

def main():
    """Fonction principale de l'exercice 2"""
    print("="*60)
    print("EXERCICE 2: STRUCTURES DE CONTRÔLE")
    print("="*60)

    # Charger les données
    donnees = charger_donnees_districts()

    print("\nListe des districts et leur classification:")
    print("-" * 50)

    # Utiliser une boucle for pour parcourir chaque district
    for index, row in donnees.iterrows():
        district = row['district']
        population = row['population']

        # Utiliser if/elif/else pour classifier selon les critères
        if population > 1000000:
            classification = "Population élevée"
        elif 500000 <= population <= 1000000:
            classification = "Population moyenne"
        else:
            classification = "Population faible"

        # Afficher les résultats
        print(f"\n{district}:")
        print(f"Population: {population:,} habitants")
        print(f"— {classification}")

if __name__ == "__main__":
    main()
```
